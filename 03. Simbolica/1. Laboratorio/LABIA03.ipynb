B. Busqueda en espacio de soluciones en Prolog


```python
ordenado([]).
ordenado([_]).
ordenado([Y,Z|L]):-Y@=<Z, ordenado([Z|L]).

mezcla([],X,X).
mezcla(X,[],X).
mezcla([Y|L1],[Z,L2],[Y|RL]):-Y@=<Z,mezcla(L1,[Z|L2],RL).
mezcla([Y|L1],[Z,L2],[Z|RL]):-Z@=<Y,mezcla(L2,[Z|L1],RL).

% Estados iniciales y terminales
initial([v, v, v, n, m, m, m]).
isTerminal([m, m, m, n, v, v, v]).

% Concatenación de listas
append([], Y, Y).
append([A | X], Y, [A | Z]) :- append(X, Y, Z).

% Acciones posibles
actionResult(EI, ES) :- append(A, [X, n | B], EI), append(A, [n, X | B], ES).
actionResult(EI, ES) :- append(A, [n, X | B], EI), append(A, [X, n | B], ES).
actionResult(EI, ES) :- append(A, [Y, X, n | B], EI), append(A, [n, X, Y | B], ES).
actionResult(EI, ES) :- append(A, [n, X, Y | B], EI), append(A, [Y, X, n | B], ES).

% Pertenece a una lista
member(E, [E | _]).
member(E, [_ | L]) :- member(E, L).

% Enlace entre estados
link(X, Y) :- actionResult(X, Y).

% Camino entre estados
path(X, Z, P, [X | Path]) :- link(X, Y), \+ member(Y, P), path(Y, Z, [Y | P], Path).
path(X, X, _, [X]).
```
